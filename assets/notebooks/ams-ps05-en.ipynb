### **Exercise 1: Shattering the Grid & Anchoring Reality (ICs and BCs)**

*Learning Objective: Understanding continuous domain formulation and the conceptual necessity of Initial and Boundary Conditions. Replacing rigid FDM meshes with random collocation points in JAX.*

A mathematical PDE like the Heat Equation is merely a rule of evolution. It tells us how heat diffuses, but mathematically, it has an infinite number of possible valid solutions. To collapse these infinite possibilities into our specific physical reality, we must anchor the mathematical fabric using two fundamental constraints:

  - **The Initial Condition (IC) –** ***The Genesis State*****:** This is the exact snapshot of the system at time  $t=0$ . It is the seed from which the timeline grows. Without an IC, the network knows the rules of physics, but doesn't know *how* our specific simulation started.
  - **The Boundary Conditions (BC) –** ***The Edges of the Sandbox*****:** These are the physical limits of our simulated world. In this case, we are simulating a 1D metal rod. The BCs dictate what happens at the extreme left and right ends of the rod across all of time (e.g., are they attached to a block of ice? Are they perfectly insulated?).

A PINN does not need a finite grid to learn this; it learns a continuous, differentiable function  $u(x, t)$ . To train it, we simply scatter random "sensor" points across the spacetime cylinder and enforce the physics at those exact coordinates.
