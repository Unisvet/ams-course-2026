### **Exercise 1: Architecting the Oracle (Subclassed Models)**

*Learning Objective: Translating mathematical theory into actionable code using Keras Subclassing APIs.*

We will not use basic, off-the-shelf Sequential models. You must build a custom bottleneck network locally before deploying to the cloud.

1.  **Sequence Windowing:** In a new script (src/architecture.py), write a function to slice your 1D signal into overlapping 2D matrices (window_size = 50 timesteps). Split the data: use the normal data (before period 60) for your training set, and the rest for testing.
2.  **The Subclassed Layer:** Using the keras.layers.Layer API, build a custom SignalCompression layer. Use a dense matrix inside the build() method to reduce the 50-timestep window down to a latent dimension of just 8. Implement the forward pass in call() using a relu activation.
3.  **The Subclassed Model:** Using the keras.Model API, build PhysicsAutoencoder, chaining your SignalCompression encoder to a SignalExpansion decoder (reconstructing 8 dimensions back to 50).
4.  **Deliverables:** Push your architecture to GitHub. *(Do not train the model locally\!)*
