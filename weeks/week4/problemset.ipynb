### **Exercise 1: The Legacy Chokehold (Sequential Simulation)**

*Learning Objective: Establishing a computational baseline and experiencing the limits of standard Python arrays and loops.*

To appreciate the speed of light, you must first experience the friction of the old world. We will simulate a massive kinetic energy harvester array consisting of **100,000 independent damped harmonic oscillators** (pendulums), each with a slightly different natural frequency.

1.  **The Sequential Swarm:** In your local uv environment, create src/legacy_swarm.py. Using standard numpy, write a pure Python function to simulate the 100,000 pendulums using a basic explicit Euler integration step over 1,000 discrete time steps.
      
      - Initialize an array of 100,000 random frequencies ( $\omega$ ).
      - Use a Python for-loop to iterate over the 1,000 time steps, updating the entire array of positions and velocities at each step.
2.  **The Stopwatch:** Use Python's time module to record the timestamp exactly before and after calling your simulation function.
3.  **Deliverables:** Note the total execution time in your terminal. Push the script to your GitHub Vault. (Prepare to wait—it will be sluggish on a standard CPU).

### **Exercise 2: The Tensor Multiverse (****vmap** **&** **jit****)**

*Learning Objective: Translating code to JAX, shattering loops with vectorization, and eliminating Python overhead via XLA compilation.*

It is time to unleash the hardware accelerators. Open Google Colab (using a GPU or TPU runtime), pull your repository, and inject uv to instantly install your dependencies.

1.  **Functional Purity:** Create src/jax_swarm.py. JAX requires "pure functions" (no side-effects, no hidden state). Import jax and jax.numpy as jnp. Write a pure mathematical function oscillator_step(x, v, w) that calculates the new state for a *single* oscillator for exactly 1 time step.
2.  **Parallel Universes (****vmap****):** Do *not* write a loop for the 100,000 frequencies\! Instead, wrap your single-oscillator function in jax.vmap. This magically transforms your scalar function into a batch-processing engine that applies the logic across the entire array of frequencies simultaneously.
3.  **Hardware Fusion (****jit****):** Wrap your outer simulation loop (which calls the vmap function) in the @jax.jit decorator. This commands the XLA compiler to fuse the mathematical operations directly into the accelerator's silicon.
4.  **The Tracing Phenomenon:** Run your JAX simulation. *Crucial:* Run it once to trigger the compilation (the "warm-up" / tracing phase), and then run it a *second time* using your timer.
5.  **Deliverables:** Calculate the **Speedup Factor** (Legacy Time / JAX 2nd Run Time). In a new markdown file docs/Ascension_Report.md, record this factor and briefly explain (1-2 sentences) *why* the first run of a JIT-compiled function is always slower than the second.

### **Exercise 3: Time Travel via Gradients (****grad****)**

*Learning Objective: Understanding Automatic Differentiation. The simulation is no longer a black box; it is a differentiable mathematical graph.*

In classical engineering, optimizing parameters (like finding the perfect initial velocity for a projectile to hit a target) requires running a simulation thousands of times using trial-and-error or numerical solvers. In the *Neo-Simulacrum*, the simulation mathematically guides you to the optimal answer automatically.

1.  **The Differentiable Target:** Write a simple pure JAX function projectile_loss(v_initial) that simulates the final distance of a thrown object after 5 seconds of flight. The function must return the Mean Squared Error (MSE) between the simulated distance and a strict target distance of exactly **150.0 meters**.
2.  **Pulling the Derivative:** Do not manually tweak v_initial. Use jax.grad(projectile_loss) to instantly create a new function that returns the *exact* analytical gradient (the slope) of the error with respect to the initial velocity.
3.  **The Optimization Loop:** Write a tiny gradient descent loop (e.g., 20 iterations). Start with a random guess  $v = 10.0$ . In each step, calculate the gradient, and update  $v$  using:  $v = v - 0.1 * gradient$ .
4.  **Deliverables:** Print the optimized  $v$  at the end of the loop. In your Ascension_Report.md, write a brief explanation of how jax.grad is fundamentally different from approximating the slope via finite differences  $\frac{f(x+h) - f(x)}{h}$ .

### **Exercise 4: Agentic Refactoring for the Horizon (Flax)**

*Learning Objective: Using an AI Agent to migrate stateful Object-Oriented architectures (Keras) into pure functional, stateless modules (Flax).*

In Week 3 with Keras 3, model weights were stored *inside* the model object itself (model.weights). JAX strictly forbids this hidden state. Therefore, Google developed **Flax**, a neural network library designed natively for JAX that mathematically separates the architecture from the weights.

1.  **Summon Observer-Prime:** Launch your Antigravity ADE or open the Gemini CLI.
2.  **The Agentic Prompt:** Feed your Agent a basic Dense layer concept and prompt it:  
    *"Observer-Prime, we are migrating our neural architecture to pure JAX. Please write a simple Python script using* *flax.linen* *to define a Multi-Layer Perceptron (MLP) module. More importantly, write a heavily commented execution block demonstrating how Flax explicitly separates model initialization (**model.init* *with a* *jax.random.PRNGKey**) from the forward pass (**model.apply**), highlighting its stateless nature compared to Keras."*
3.  **Deliverables:** Save the AI-generated Flax code in src/flax_core.py. Run it to ensure it executes. Paste the Agent's explanation regarding explicit state management (Flax) vs. implicit state (Keras) into your Ascension_Report.md and push your Vault.
